## Encoding :-

```python
from sklearn.preprocessing import LabelEncoder,OneHotEncoder,OrdinalEncoder
```

```python
sex_encoder =LabelEncoder()
age=LabelEncoder()
genhealth=LabelEncoder()
phyacti=LabelEncoder()
smokestat=LabelEncoder()
alcodrink=LabelEncoder()
sleephrs=LabelEncoder()
hadstrk=LabelEncoder()
hadcopd=LabelEncoder()
hadangina=LabelEncoder()
haddia=LabelEncoder()
bmi=LabelEncoder()
dfwalk=LabelEncoder()
```

```python
sexv=x['Sex']
agee=x['AgeCategory']
genhealthe=x['GeneralHealth']
phyactie=x['PhysicalActivities']
smokestate=x['SmokerStatus']
alcodrinke=x['AlcoholDrinkers']
sleephrse=x['SleepHours']
hadstrke=x['HadStroke']
hadcopde=x['HadCOPD']
hadanginae=x['HadAngina']
haddiae=x['HadDiabetes']
bmie=x['BMI']
dfwalke=x['DifficultyWalking']
```

```python
sex_encoded=sex_encoder.fit_transform(sexv)
age_encoded=age.fit_transform(agee)
genhealth_encoded=genhealth.fit_transform(genhealthe)
phyactie_encoded=phyacti.fit_transform(phyactie)
smokestate_encoded=smokestat.fit_transform(smokestate)
alcoholdrick_encoded=alcodrink.fit_transform(alcodrinke)
sleephrs_encoded=sleephrs.fit_transform(sleephrse)
hadstroke_encoded=hadstrk.fit_transform(hadstrke)
hadcopd_encoded=hadcopd.fit_transform(hadstrke)
hadangin_encoded=hadangina.fit_transform(hadanginae)
haddia_encoded=haddia.fit_transform(haddiae)
bmie_encoded=bmi.fit_transform(bmie)
dfwalke_encoded=dfwalk.fit_transform(dfwalke)
```

```python
sex_encoded
```

```python
x['Sex']=sex_encoded
x['AgeCategory']=age_encoded
x['GeneralHealth']=genhealth_encoded
x['PhysicalActivities']=phyactie_encoded
x['SmokerStatus']=smokestate_encoded
x['AlcoholDrinkers']=alcoholdrick_encoded
x['SleepHours']=sleephrs_encoded
x['HadStroke']=hadstroke_encoded
x['HadCOPD']=hadcopd_encoded
x['HadAngina']=hadangin_encoded
x['HadDiabetes']=haddia_encoded
x['BMI']=bmie_encoded
x['DifficultyWalking']=dfwalke_encoded
```

```python
x.head(10)
```

```python
y = df['HadHeartAttack']
y.head(5)
```

```python
ye = LabelEncoder()
```

```python
ydata=ye.fit_transform(y)
ydata
```

## Scaling :-

```python
from sklearn.preprocessing import StandardScaler
```

```python
scaler = StandardScaler()
```

```python
xscaled=scaler.fit_transform(x)
```

```python
xscaled
```

## Train Test Split :-

```python
from sklearn.model_selection import train_test_split
```

```python
xtrain,xtest,ytrain,ytest=train_test_split(xscaled,ydata,random_state=42,test_size=0.2)
```

## Model Building :-

```python
from sklearn.linear_model import LogisticRegression
```

```python
model = LogisticRegression()
```

```python
model
```

```python
model.fit(xtrain,ytrain)
```

```python
pred = model.predict(xtest)
```

```python
model.coef_
```

```python
model.intercept_
```

## Evaluation Metrics :-

```python
from sklearn.metrics import accuracy_score,confusion_matrix,precision_score,recall_score,f1_score,roc_curve,auc
```

```python
accuracy_score(ytest,pred)
```

```python
confusion_matrix(ytest,pred)
```

```python
precision_score(ytest,pred)
```

```python
recall_score(ytest,pred)
```

```python
f1_score(ytest,pred)
```

## roc,auc curve :-

```python
proba = model.predict_proba(xtest) 
```

```python
proba
```

```python
import matplotlib.pyplot as plt
```

```python
# fpr, tpr, thresholds = roc_curve(ytest, proba[:, 1])  # Take the second column (probability of positive class)

# auc(fpr,tpr)

# plt.line(fpr,tpr)
# plt.show()
```

```python
import matplotlib.pyplot as plt
from sklearn.metrics import roc_curve, auc

# Assuming you already have fpr, tpr, and thresholds from roc_curve
fpr, tpr, thresholds = roc_curve(ytest, proba[:, 1])  # Take the second column (probability of positive class)

# Calculate AUC
roc_auc = auc(fpr, tpr)

# Plot the ROC curve
plt.plot(fpr, tpr, label=f'ROC curve (area = {roc_auc:.2f})')
plt.plot([0, 1], [0, 1], 'k--')  # Diagonal line for random classifier
plt.xlim([0.0, 1.0])
plt.ylim([0.0, 1.05])
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.title('Receiver Operating Characteristic (ROC) Curve')
plt.legend(loc='lower right')
plt.show()
```
